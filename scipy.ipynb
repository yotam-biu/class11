## Minimum of a Function

```python
from scipy.optimize import minimize

my_func = lambda x: x**x
result = minimize(my_func, 0.5)
minimum_x = result.x[0]
minimum_y = result.fun

print(f"Minimum of x^x occurs at x = {minimum_x}, f(x) = {minimum_y}")
```

# 4. Root
  $\tanh(\arctan(x))$



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root
import sympy as sp

# Define the function f(x)
def f(x):
    return np.tanh(np.arctan(x))

# First derivative using finite differences
def first_derivative(f, x, dx=1e-6):
    return (f(x + dx) - f(x - dx)) / (2 * dx)

# Second derivative using finite differences
def second_derivative(f, x, dx=1e-6):
    return (f(x + dx) - 2 * f(x) + f(x - dx)) / (dx**2)

# Find the numeric inflection point
def second_derivative_zero(x):
    return second_derivative(f, x)


x_vals = np.linspace(-5, 5, 500)
y_vals = f(x_vals)
first_derivative_vals = [first_derivative(f, x) for x in x_vals]
second_derivative_vals = [second_derivative(f, x) for x in x_vals]

plt.plot(x_vals, y_vals, label=r"$f(x) = \tanh(\arctan(x))$", color='blue')
plt.plot(x_vals, first_derivative_vals, label=r"$f'(x)$", color='green')
plt.plot(x_vals, second_derivative_vals, label=r"$f''(x)$", color='red')

plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend();
```

```python
import numpy as np
from scipy.optimize import root

def f(x):
    return np.tanh(np.arctan(x))

dx=1e-6
def dfdx(x):
    return (f(x + dx) - f(x-dx)) / (2*dx)
def df2dx2(x):
    return (f(x + dx) - 2 * f(x) + f(x - dx)) / (dx**2)

inflection_point_result = root(df2dx2, 0)
inflection_point = inflection_point_result.x[0]

print(f"Inflection point (numeric): x = {inflection_point}")
```

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd

R = 8.314  # Universal gas constant in J/molK
def clausius_clapeyron(T, delta_H, C):
    return -delta_H / (R * T) + C

df = pd.read_csv("clausius_clapeyron_data.csv")
T = df['T']
lnP = df['lnP']
popt, _ = curve_fit(clausius_clapeyron, T, lnP)
delta_H_fit, C_fit = popt

plt.scatter(T, lnP, color='blue')
plt.plot(T, clausius_clapeyron(T, delta_H_fit, C_fit), color='black')
plt.xlabel('T', fontsize=16)
plt.ylabel('ln(P)', fontsize=16);
```

```python
import matplotlib.pyplot as plt

plt.scatter(1 / T, lnP, color='blue')
plt.plot(1 / T, clausius_clapeyron(T, delta_H_fit, C_fit), color='black')
plt.xlabel('1/T', fontsize=16)
plt.ylabel('ln(P)', fontsize=16);
```
